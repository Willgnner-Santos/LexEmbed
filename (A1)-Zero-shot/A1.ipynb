```python
import re

def remove_pontuacao_inicio(text):
    if isinstance(text, str):
        return re.sub(r'^[^\w\s]+', '', text)
    return text

# Function to replace punctuation repeated more than 4 times with only 4
def remove_pontuacao_repetida(text):
    if isinstance(text, str):
        return re.sub(r'([^\w\s])\1{4,}', r'\1\1\1\1', text)
    return text
```

```python
import pandas as pd

df = pd.read_csv(r'/workspace/src/data/dados_berna_comparacao_fato_direito_04082025.csv', encoding='utf-8')
df['processo_paradigma'].value_counts()
```

```python
cols = ['fato_paradigma', 'fato_similar', 'direito_paradigma',"direito_similar"]

df[cols] = df[cols].replace({r'[\r\n]+': ' '}, regex=True)

df[cols] = df[cols].map(remove_pontuacao_inicio)

df = df.map(lambda x: x.strip() if isinstance(x, str) else x)

df[cols] = df[cols].map(remove_pontuacao_repetida)
```

```python
df_5_fatos = (
    df.drop_duplicates(subset=['fato_paradigma', 'fato_similar'])
      .groupby('fato_paradigma')
      .head(5)
      .reset_index(drop=True)
)
df = df_5_fatos.head(500)
df
```

```python
df_unique = df.drop_duplicates(subset=['fato_similar','direito_similar'])

print(f"Total de linhas únicas: {len(df_unique)}")
# Grouping and selecting, for example, up to 5 unique similar facts per process.
df = (
    df_unique.groupby('processo_paradigma')
                .apply(lambda x: x)
             .reset_index(drop=True)
)
```

```python
import pandas as pd

# Unifies fact and thesis for paradigm and similar purposes.
df["paradigma_unificado"] = df["fato_paradigma"].astype(str) + " " + df["direito_paradigma"].astype(str)
df["similar_unificado"] = df["fato_similar"].astype(str) + " " + df["direito_similar"].astype(str)

# Normalizes the similarity for the range [0, 1]
sim_min = df["similaridade_fato"].min()
sim_max = df["similaridade_fato"].max()
df["similaridade_normalizada"] = (df["similaridade_fato"]  + df["similaridade_tese"]) / 2

# Creates the new DataFrame for pairs
df = pd.DataFrame({
    "processo_paradigma": df["processo_paradigma"],
    "processo_similar": df["processo_similar"],
    "sentence1": df["paradigma_unificado"],
    "sentence2": df["similar_unificado"],
    "score": df["similaridade_normalizada"]
})

df
```

```python
sentences =  df['sentence2'].tolist()
```

```python
from sentence_transformers import SentenceTransformer, util
# Model for computing sentence embeddings. 
model = SentenceTransformer(r'/workspace/Similaridade-Fato-Tese/modelo_ft_similaridade',trust_remote_code=True)

embeddings = model.encode(sentences, show_progress_bar= True, batch_size=16)
```

```python
from sklearn.preprocessing import normalize
from sklearn.metrics import silhouette_score
import numpy as np
import sklearn.cluster as cluster
import hdbscan
import umap
import matplotlib.pyplot as plt
import seaborn as sns

# Reduces embeddings to 2D for visualization
reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, metric='cosine', random_state=42)
reduced_embeddings = reducer.fit_transform(embeddings)

# Função para rodar clustering e # Function to run clustering and plotting
def run_clustering_and_score(algorithm_class, data, reduced_data_2d, algorithm_args=(), algorithm_kwargs={}, metric='cosine'):
    try:
        model = algorithm_class(*algorithm_args, **algorithm_kwargs)
        labels = model.fit_predict(data)
        
        # Check if there is more than 1 cluster
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        if n_clusters <= 1:
            print(f"{algorithm_class.__name__}: Menos de 2 clusters encontrados. Silhouette Score não calculado.")
        else:
            score = silhouette_score(data, labels, metric=metric)
            print(f"{algorithm_class.__name__}: Silhouette Score = {score:.4f} (clusters: {n_clusters})")

        # Plot of the clusters
        #plot_clusters_2d(reduced_data_2d, labels, algorithm_class.__name__)
        
    except Exception as e:
        print(f"{algorithm_class.__name__}: Erro - {str(e)}")

# Function to plot clusters with colors
def plot_clusters_2d(data_2d, labels, title):
    plt.figure(figsize=(8, 5))
    unique_labels = np.unique(labels)
    palette = sns.color_palette('hls', len(unique_labels))
    colors = [palette[label % len(palette)] if label >= 0 else (0.5, 0.5, 0.5) for label in labels]
    plt.scatter(data_2d[:, 0], data_2d[:, 1], c=colors, s=10, alpha=0.7)
    plt.title(f"{title} - Cluster Plot")
    plt.xlabel("UMAP Dim 1")
    plt.ylabel("UMAP Dim 2")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Normalization for HDBSCAN
norm_embeddings = normalize(embeddings, norm='l2')

# Data by algorithm
embedding_options = {
    'KMeans': embeddings,
    'AgglomerativeClustering': embeddings,
    'SpectralClustering': embeddings,
    'AffinityPropagation': embeddings,
    'DBSCAN': embeddings,
    'HDBSCAN': norm_embeddings
}

# List of algorithms to test
algorithms = [
    (cluster.KMeans, {'n_clusters': 300, 'random_state': 42}),
    (cluster.AgglomerativeClustering, {'n_clusters': 300}),
    (cluster.SpectralClustering, {'n_clusters': 300, 'affinity': 'nearest_neighbors'}),
    (cluster.AffinityPropagation, {'preference': -5.0, 'damping': 0.95}),
    (cluster.DBSCAN, {'eps': 0.025, 'metric': 'cosine'}),
    (hdbscan.HDBSCAN, {})
]

# Rotate and plot all
for algo_class, params in algorithms:
    data = embedding_options.get(algo_class.__name__, embeddings)
    run_clustering_and_score(algo_class, data, reduced_embeddings, algorithm_kwargs=params)
```
